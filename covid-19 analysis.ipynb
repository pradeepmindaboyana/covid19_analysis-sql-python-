```python
import zipfile
zip_ref = zipfile.ZipFile('Covid%20Data.csv.zip') 
zip_ref.extractall() # extract file to dir
zip_ref.close() # close file
```

```python
#read data from the file and handle null values
import pandas as pd
df = pd.read_csv("covid data.csv",na_values=[97,99,98]) #values as 97,98 and 99 are missing data, so we are replacing them with NaN.In this case, any occurrence of 97 or 99 in the CSV file will be treated as a missing value (NaN) in the resulting DataFrame.
df
```

```python
#rename columns names ..make them lower case.
df.columns # to print columns in the csv file
df.columns=df.columns.str.lower()  # we are converting column names to lower case and asssigning them to columns 
df.columns # printing columns
df.dtypes # printing datatypes

#Note:Object dtype is the default data type for text (string) data in pandas. It can also handle NaN values.
```

```python
#in the covid dataset some column values representing 1 -'YES',2 -'No'. So we are replacing those 1,2 values with the respective str. 

# Step1 :selecting all columns that we need to replace the row values..
columns_to_replace = ['intubed', 'pneumonia','pregnant','diabetes','asthma','inmsupr','hipertension','cardiovascular','obesity','renal_chronic','tobacco','icu','other_disease','copd']

#Step2 :defining the values and corresponding replace value.
df[columns_to_replace] = df[columns_to_replace].replace({1:'Yes',2:'No'}).astype(str)

#step3 : printing modified DataFrame
df
```

```python
#Changing data types of columns, converting float values to integer values

#fill NaN values with zero
df['age'] = df['age'].fillna(0)

#convert age column float dtype to integer
df['age'] = df['age'].astype(int)

#print DataFrame 
df
```

```python
#modifying classification final column.
df['clasiffication_final']=df['clasiffication_final'].replace({1:'Mild_degree_cvpostive',2:'Moderate_degree_cvpostive',3:'severe_degree_cvpostive',4:'cvnegative'})
df['clasiffication_final'] = df['clasiffication_final'].apply(
    lambda x: 'Undetermined/unconclusive' if x in [5, 6, 7] else x
)
df
```

```python
# replacing sex column 
df['sex']=df['sex'].replace({1:'Female',2:'Male'})

# replacing patient_type column
df['patient_type']=df['patient_type'].replace({1:'Returned_Home',2:'Hospitalization'})

df
```

```python
#load the data into sql server using replace option
#!pip install sqlalchemy --it is a library used to connect with sql server.SQLAlchemy is a powerful and flexible Python library used for working with relational databases
import sqlalchemy as sal
#This line creates an SQLAlchemy engine object, which represents the core interface to the database.

# mssql://: Specifies the dialect. In this case, mssql indicates that you are connecting to a Microsoft SQL Server database.
#PRADEEP\SQLEXPRESS: Specifies the server name or instance of SQL Server. Here, PRADEEP is the machine name, and SQLEXPRESS is the named instance of SQL Server.
#/Analytics: Indicates the name of the database you want to connect to, in this case, Analytics.
#?driver=ODBC+DRIVER+17+FOR+SQL+SERVER: Specifies the ODBC driver to be used for the connection. ODBC DRIVER 17 FOR SQL SERVER is a specific driver that enables communication between Python and SQL Server.

engine = sal.create_engine('mssql://PRADEEP\SQLEXPRESS/Analytics?driver=ODBC+DRIVER+17+FOR+SQL+SERVER')

#This line opens a connection to the database using the engine you just created.
conn=engine.connect()
```

```python
#load the data into sql server using append option

df.to_sql('df_covid19', con=conn , index=False, if_exists = 'replace')
```
